```python
import pandas as pd

# Supposons que merged_data est votre DataFrame fusionné contenant les notes
# Normalisation des colonnes sur une échelle de 0 à 5

# Normalisation des notes de RottenTomatoes et RottenTomatoes_User (valeurs maximales 100)
merged_data['RottenTomatoes_Normalized'] = merged_data['RottenTomatoes'] / 20
merged_data['RottenTomatoes_User_Normalized'] = merged_data['RottenTomatoes_User'] / 20

# Normalisation des notes de Metacritic et Metacritic_User (valeurs maximales 100)
merged_data['Metacritic_Normalized'] = merged_data['Metacritic'] / 20
merged_data['Metacritic_User_Normalized'] = merged_data['Metacritic_User'] / 2  # Valeur max 10

# Normalisation des notes IMDb (valeurs maximales 10)
merged_data['IMDB_Normalized'] = merged_data['IMDB'] / 2  # Valeur max 10

# Afficher les résultats de la normalisation
print(merged_data[['FILM', 'RottenTomatoes', 'RottenTomatoes_User', 
                   'Metacritic', 'Metacritic_User', 'IMDB', 
                   'RottenTomatoes_Normalized', 'RottenTomatoes_User_Normalized', 
                   'Metacritic_Normalized', 'Metacritic_User_Normalized', 
                   'IMDB_Normalized']].head())
```
